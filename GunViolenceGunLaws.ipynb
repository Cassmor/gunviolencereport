Next I added chloropeth maps to folium base maps. For this map, geo_data was my us_states_id2STATE2 and my data was Median_H_Income_2022_df. The choropleth map can join the data using the key_on argument. Both data sets have a column holding state names, so I used name for the geo_data for the key since it holds the names of the States, to add a tooltip, so a user can hover over a state and get information about the data from columns of the specified data frame. For all the maps, I use the geo-data file (since the function needs the goegraphic data), I just chose different columns to supply the relevant information.
